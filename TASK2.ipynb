```python
from google.colab import files
uploaded = files.upload()
```

```python
import csv
from fpdf import FPDF

# Automatically use the uploaded file name
filename = next(iter(uploaded))

names = []
scores = []

with open(filename, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        names.append(row['Name'])
        scores.append(int(row['Score']))

average_score = sum(scores) / len(scores)

# Create PDF
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)

pdf.set_text_color(0, 0, 128)
pdf.cell(200, 10, txt="Student Score Report", ln=True, align='C')
pdf.ln(10)

pdf.set_text_color(0, 0, 0)
pdf.cell(100, 10, "Name", 1)
pdf.cell(50, 10, "Score", 1)
pdf.ln()

for name, score in zip(names, scores):
    pdf.cell(100, 10, name, 1)
    pdf.cell(50, 10, str(score), 1)
    pdf.ln()

pdf.ln(10)
pdf.set_text_color(0, 128, 0)
pdf.cell(200, 10, f"Average Score: {average_score:.2f}", ln=True)

pdf_file = "sample_report.pdf"
pdf.output(pdf_file)

print(f"✅ PDF Report Generated: {pdf_file}")
```
