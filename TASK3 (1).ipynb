```python
# ai_chatbot_nltk.py

import nltk
from nltk.chat.util import Chat, reflections

nltk.download('punkt')

chatbot_pairs = [
    [
        r"hi|hello|hey",
        ["Hello! How can I help you today?", "Hi there! Need any assistance?"]
    ],
    [
        r"what is your name?",
        ["I am a CodTech Internship Chatbot.", "You can call me CodBot."]
    ],
    [
        r"how are you ?",
        ["I'm just a bot, but I'm doing great! How can I assist you?"]
    ],
    [
        r"(.*) your creator?",
        ["I was created by a CodTech intern using Python and NLTK."]
    ],
    [
        r"(.*) (location|city) ?",
        ["I'm virtual, but I'm here to help you wherever you are!"]
    ],
    [
        r"(.*) help (.*)",
        ["Of course! I'm here to assist you. Please tell me your query."]
    ],
    [
        r"quit",
        ["Goodbye! Have a great day!"]
    ]
]

chatbot = Chat(chatbot_pairs, reflections)

def start_chat():
    print("Welcome to the CodTech AI Chatbot! (Type 'quit' to exit)")
    chatbot.converse()

if __name__ == "__main__":
    start_chat()
```
